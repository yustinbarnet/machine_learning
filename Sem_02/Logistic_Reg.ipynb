### In Synthetic data

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
````

````python
x, y = make_classification(n_samples = 250 , n_features = 2 , n_redundant = 0, n_informative = 2, n_clusters_per_class = 1, n_classes = 2, random_state = 3)

print(x.shape, y.shape)

#realizacion de diagrama de dispersion
_, ax = plt.subplots(figsize=( 5 , 4 )) #(ancho,alto)
ax.scatter( x[:, 0] , x[:, 1] , c = y , cmap = 'Paired')
````

````python
# Create an instance
logistic_regression = LogisticRegression()

# Train the model
logistic_regression.fit ( x , y ) #(# Samples, # Features)
````

````python
print(logistic_regression.coef_,logistic_regression.intercept_)
````

### **Decision boundaries**

````python
# Min-Max values for the grid of decision boundaries
h_min, h_max = x[:, 0].min()-0.1, x[:,0].max()+0.1  #eje horizontal
v_min, v_max = x[:, 1].min()-0.1 , x[:, 1].max()+0.1 #eje vertical

# Create grid
h_grid, v_grid = np.meshgrid( np.linspace(h_min,h_max,500) ,  np.linspace(v_min,v_max,500) )

print(h_grid.shape,v_grid.shape,h_grid.ravel().shape,v_grid.ravel().shape)
print(np.c_[h_grid.ravel(),v_grid.ravel()].shape)

# Estimate predictions over the grid
predictions = logistic_regression.predict(np.c_[h_grid.ravel(), v_grid.ravel()])
print (predictions.shape)
predictions = predictions.reshape(h_grid.shape)
print (predictions.shape)
````

````python

# Draw grid
_, ax = plt.subplots (figsize=(7,4))
ax.pcolormesh( h_grid , v_grid , predictions , cmap = "Paired")

# Scatter real values
ax.scatter ( x[:,0] , x[:,1] , c = y , edgecolors = 'k', cmap = "Paired")

#ax.set_xticks(())
#ax.set_yticks(())
````

### In Real data

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
````

````python
iris = load_iris() # Load iris dataset
iris_x = iris.data[:, 0:2] # Features
iris_y = iris.target # Labels

# Imprimir formas
print (iris_x.shape, iris_y.shape)
````

````python
# Create an instance
decision_tree = LogisticRegression()

# Train the model
decision_tree.fit ( iris_x , iris_y ) #(# Samples, # Features)
````

````python
#realizacion de diagrama de dispersion
_, ax = plt.subplots(figsize=( 5 , 4 )) #(ancho,alto)
ax.scatter( iris_x[:, 0] , iris_x[:, 1] , c = iris_y , cmap = 'Paired')
````

````python
# Min-Max values for the grid of decision boundaries
x_min, x_max = iris_x[:, 0].min()-0.1, iris_x[:,0].max()+0.1
y_min, y_max = iris_x[:, 1].min()-0.1 , iris_x[:, 1].max()+0.1

# Create grid
x_grid, y_grid = np.meshgrid( np.linspace(x_min,x_max,500) ,  np.linspace(y_min,y_max,500))

# Estimate predictions over the grid
predictions = decision_tree.predict(np.c_[x_grid.ravel(), y_grid.ravel()])
print (predictions.shape)
````

````python
predictions = predictions.reshape(x_grid.shape)
print (predictions.shape)
# Draw grid
_, ax = plt.subplots (figsize=(8,5))
ax.pcolormesh( x_grid , y_grid , predictions , cmap = "Paired")

# Scatter real values
ax.scatter( iris_x[:, 0] , iris_x[:, 1] , c = iris_y ,edgecolors = 'k', cmap = 'Paired')

#ax.set_xticks(())
#ax.set_yticks(())
````
