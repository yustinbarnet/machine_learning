# **Density-based spatial clustering (DBSCAN)**

DBSCAN documentation [here](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.DBSCAN.html)

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
```

```python
X, _ = make_blobs(n_samples = 300, n_features = 2, random_state=2)

# Train test splits
X_train, X_test=  train_test_split(X, test_size=0.2, random_state=1)


_, ax = plt.subplots(figsize = (5,5))
ax.scatter(X_train[:,0], X_train[:,1], c='b', edgecolors='k')
ax.scatter(X_test[:,0], X_test[:,1], c='r', edgecolors='k')

# plt.show()
```

```python
dbscan = DBSCAN(eps= 1.4, min_samples= 4 ) #el valor de estos es con base al valor del rango (dimension) de los datos, esto se ve egracias a las graficas y un porcentaje

# Train DBSCAN
predictions= dbscan.fit_predict(X_test)
```

```python
_, ax = plt.subplots(figsize = (5,4))
ax.scatter( X_test[:,0] , X_test[:,1] , c= predictions , edgecolors='k', cmap='Paired')
```

```python
#Conocer la cantidad de elementos en cada prediccion
print(np.unique(predictions, return_counts=True))

#dbscan utiliza el -1 para saber los outliers
```

# **DBSCAN on real data**

Download the fruits data [here](https://drive.google.com/file/d/1M8tiAWDZclABJN1Meq9oEHCLJTNyHdpP/view?usp=share_link)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
# Cargar los datos
data_path = '/content/fruits_data_ext.csv'
data = pd.read_csv(data_path)

# Seleccionar las caractersticas para el clustering
X = data[['width', 'height']].values

# Visualizacin de los datos
plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], edgecolors='k')
plt.show()
```

```python
# Configuracin y entrenamiento del modelo DBSCAN
eps_value = 2.0  # Este valor puede ajustarse segn la distribucin de los datos
min_samples_value = 5
dbscan = DBSCAN(eps=eps_value, min_samples=min_samples_value)
dbscan.fit(X)

# Predicciones de clusters y deteccin de outliers
predictions = dbscan.labels_  # -1 indica outliers

# Visualizacin de clusters y outliers
plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], c=predictions, cmap='Paired', edgecolors='k')
plt.title(f'Clustering de frutas usando DBSCAN (eps={eps_value}, min_samples={min_samples_value})')
plt.colorbar(label='Cluster')
plt.show()

# Resumen de clusters y outliers
unique, counts = np.unique(predictions, return_counts=True)
print("Nmero de elementos en cada cluster:", dict(zip(unique, counts)))
```
