# **Linear Regresion**

```linear_regression``` module documentation [here](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html?highlight=linear%20regression#sklearn.linear_model.LinearRegression)

````python
import numpy as np # Numerical library
import matplotlib.pyplot as plt # plots
from sklearn.linear_model import LinearRegression
````

````python
def f(x) : # Create a function that samples from linear function
  y = (0.0918 * x) + 1.2859 + 0.1*np.random.randn (x.shape[0]) # Linear function
  return y

#0.1*np.random.randn (x.shape[0]) esto es el factor de ruido

# create training samples
x = np.arange (0, 20, 0.25); print (x.shape)

# Generate samples
y = f(x) ; print (y.shape)

# Scatter plot
_,ax= plt.subplots(figsize=(5,4))
ax.scatter(x, y, color = 'blue')
````

````python
# Create an instance
linear_regression = LinearRegression()

# Train the model
linear_regression.fit( x.reshape(-1,1) , y ) # (# Samples, # Features)
````

````python
# Print estimated values
print ("w = " + str(linear_regression.coef_) + ", b = " + str(linear_regression.intercept_)) #coef=pendiente intercept=intercepto
````

````python
# Create a new sample estimate prediction
new_sample = np.array([5])
print (f(new_sample))

prediction = linear_regression.predict( new_sample.reshape(-1,1) )
print (prediction)
````

````python
# Predict x values
predictions = linear_regression.predict( x.reshape(-1,1) )

# Draw real samples (blue) and predictions (red)
_, ax = plt.subplots (figsize = ( 5 , 4 ))
ax.scatter( x , y , color = 'blue')
ax.plot ( x , predictions , color = 'red') #para graficar una linea
````

## In Synthetic data

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
````

````python
X, y = make_regression(n_samples = 100 , n_features = 1, noise = 10, random_state = 5 )
_, ax = plt.subplots (figsize=( 5 , 4 ))
ax.scatter( X , y , s=10,c = 'b')
print (X.shape, y.shape)
````

````python
linear_regression = LinearRegression()
linear_regression.fit( X.reshape(-1,1) , y )
````

````python
# Imprimir los valores estimados para compararlos respecto a los originales
print ("w = " + str(linear_regression.coef_) + ", b = " + str(linear_regression.intercept_))
````

````python
# Predecir la nueva muestra. Imprimir la prediccion
prediction = linear_regression.predict( X.reshape(-1,1) )
print (prediction.shape)
````

````python
# Draw real samples (blue) and predictions (red)
_, ax = plt.subplots (figsize = ( 5 , 4 ))
ax.scatter( X , y , s=10, color = 'blue')
ax.plot ( X , prediction , color = 'g',linewidth=1) #para graficar una linea
````
