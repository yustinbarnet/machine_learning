## On real data - Fruits dataset

Download the fruits data [here](https://drive.google.com/file/d/1M8tiAWDZclABJN1Meq9oEHCLJTNyHdpP/view?usp=share_link)

````
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
````

````
fruit_data = pd.read_csv('/content/fruits_data_ext.csv')
fruits_x = fruit_data[['width', 'height']].values  # Features: 'width' and 'height'
fruits_y = fruit_data['fruit_label'].values  # Labels: 'fruit_label'

# Convert everything to a np array
fruits_x, fruits_y = np.asarray(fruits_x), np.asarray(fruits_y)

# Print shapes
print(fruits_x.shape, fruits_y.shape)
````

````
# Instance GaussianNB object
naive_bayes = GaussianNB()

# Training
naive_bayes.fit(fruits_x, fruits_y)
````

````
## Plot decision boundaries
# Paso 1: Determinar los valores mínimo y máximo para el grid
h_min, h_max = fruits_x[:, 0].min() - 1, fruits_x[:, 0].max() + 1
v_min, v_max = fruits_x[:, 1].min() - 1, fruits_x[:, 1].max() + 1

# Paso 2: Crear una cuadrícula de puntos
h_grid, v_grid = np.meshgrid(np.linspace(h_min, h_max, 200), np.linspace(v_min, v_max, 200))

# Paso 3: Predecir sobre la cuadrícula
pred_grid = naive_bayes.predict(np.c_[h_grid.ravel(), v_grid.ravel()])

# Paso 4: Cambiar el tamaño de las predicciones para ajustarlo al grid
pred_grid = pred_grid.reshape(h_grid.shape)

# Paso 5: Crear el gráfico
_, ax = plt.subplots(figsize=(5, 4))  # Establecer tamaño de la gráfica
ax.pcolormesh(h_grid, v_grid, pred_grid, cmap='Paired', shading='auto')  # Malla con fronteras de decisión
ax.scatter(fruits_x[:, 0], fruits_x[:, 1], c=fruits_y, edgecolors='k', cmap='Paired')  # Puntos de los datos


plt.show()
````
